## Bar Size / Granularity

```python
from pandas.tseries.frequencies import to_offset
```

```python
ada
```

```python
ada.loc["08-2020"].iplot(kind= "candle")
```

```python
weekly = ada.resample("W-Fri").ohlc()
weekly
```

```python
weekly = ada.resample("W-Fri").agg({"Open":"first",
                                   "High":"max",
                                   "Low":"min",
                                   "Close":"last"
                                  })
weekly
```

```python
weekly.index = weekly.index - to_offset("4d")
```

```python
weekly.loc["08-2020":"08-2021"].iplot(kind= "candle")
```
